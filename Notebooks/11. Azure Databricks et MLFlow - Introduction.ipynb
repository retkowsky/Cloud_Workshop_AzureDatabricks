### Il est nécessaire de créer dans le Workspace Shared un dossier qui contiendra les restitutions MLFlow.
Pour cela création d'un dossier **experiments** dans le dossier partagé Azure Databricks **("/Shared/experiments")**

```
# Connection MLFlow et création expérience
import mlflow
mlflow.set_experiment("/Shared/experiments/Test")
```

```
# Démarrage run MLFlow 1

with mlflow.start_run():
  # Log a parameter (key-value pair)
  mlflow.log_param("param1", 5)

  # Log a metric; metrics can be updated throughout the run
  mlflow.log_metric("Métrique 1", 1)
  mlflow.log_metric("Métrique 2", 2)
  mlflow.log_metric("Métrique 3", 3)
  mlflow.log_metric("Métrique 4", 4)
  mlflow.log_metric("Métrique 5", 5)

  # Log an artifact (output file)
  with open("output.txt", "w") as f:
      f.write("Hello. Simulation 1")
  mlflow.log_artifact("output.txt")
```

```
# Démarrage run MLFlow 2

with mlflow.start_run():
  # Log a parameter (key-value pair)
  mlflow.log_param("param1", 10)

  # Log a metric; metrics can be updated throughout the run
  mlflow.log_metric("Métrique 1", 10)
  mlflow.log_metric("Métrique 2", 20)
  mlflow.log_metric("Métrique 3", 30)
  mlflow.log_metric("Métrique 4", 40)
  mlflow.log_metric("Métrique 5", 50)

  # Log an artifact (output file)
  with open("output.txt", "w") as f:
      f.write("Hello. Simulation 2")
  mlflow.log_artifact("output.txt")
```
